# Data Visualization

Creating visualizations with matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Generating Data

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

print(f'Generated {len(x)} data points')
print(f'X range: [{x.min():.2f}, {x.max():.2f}]')
print(f'Y range: [{y.min():.2f}, {y.max():.2f}]')
```

```python
statistics = {
    'mean': y.mean(),
    'std': y.std(),
    'min': y.min(),
    'max': y.max()
}
statistics
```
