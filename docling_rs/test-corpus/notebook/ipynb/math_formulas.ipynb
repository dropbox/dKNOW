## Calculating Pi

Using the Leibniz formula approximation.

```python
def calculate_pi(n):
    pi_approx = 0
    for i in range(n):
        pi_approx += ((-1)**i) / (2*i + 1)
    return pi_approx * 4

pi_value = calculate_pi(10000)
print(f'Approximation of Pi: {pi_value}')
print(f'Actual Pi value: {math.pi}')
print(f'Difference: {abs(pi_value - math.pi)}')
```
